# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
data = pd.read_csv('Position_Salaries.csv')

X = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values

print(X)
print(y)
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression

regressor_lin = LinearRegression()

regressor_lin.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures

regressor_poly = PolynomialFeatures(degree=2)

X_poly = regressor_poly.fit_transform(X)

regressor_lin_2 = LinearRegression()

regressor_lin_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```
plt.scatter(X, y, color='red')

plt.plot(X, regressor_lin.predict(X), color='blue')

plt.title('Truth or Bluff (Linear Regression)')

plt.xlabel('Position Level')

plt.ylabel('Salary')

plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, y, color='red')

plt.plot(X, regressor_lin_2.predict(X_poly), color='blue')

plt.title('Truth or Bluff (Linear Regression)')

plt.xlabel('Position Level')

plt.ylabel('Salary')

plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
regressor_poly_3 = PolynomialFeatures(degree=3)

X_poly_3 = regressor_poly_3.fit_transform(X)

regressor_lin_3 = LinearRegression()

regressor_lin_3.fit(X_poly_3, y)

plt.scatter(X, y, color='red')

plt.plot(X, regressor_lin_3.predict(X_poly_3), color='blue')

plt.title('Truth or Bluff (Linear Regression)')

plt.xlabel('Position Level')

plt.ylabel('Salary')

plt.show()
```

```
regressor_poly_4 = PolynomialFeatures(degree=4)

X_poly_4 = regressor_poly_4.fit_transform(X)

regressor_lin_4 = LinearRegression()

regressor_lin_4.fit(X_poly_4, y)

plt.scatter(X, y, color='red')

plt.plot(X, regressor_lin_4.predict(X_poly_4), color='blue')

plt.title('Truth or Bluff (Linear Regression)')

plt.xlabel('Position Level')

plt.ylabel('Salary')

plt.show()
```

```
regressor_poly_100 = PolynomialFeatures(degree=100)

X_poly_100 = regressor_poly_100.fit_transform(X)

regressor_lin_100 = LinearRegression()

regressor_lin_100.fit(X_poly_100, y)

plt.scatter(X, y, color='red')

plt.plot(X, regressor_lin_3.predict(X_poly_3), color='black')
plt.plot(X, regressor_lin_4.predict(X_poly_4), color='green')
plt.plot(X, regressor_lin_100.predict(X_poly_100), color='green')

plt.title('Truth or Bluff (Linear Regression)')

plt.xlabel('Position Level')

plt.ylabel('Salary')

plt.show()
```
